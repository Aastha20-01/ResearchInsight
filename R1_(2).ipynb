```python
import os
import fitz

pdf_folder = "pdf"
text_folder = "texts"

os.makedirs(text_folder, exist_ok=True)

for filename in os.listdir(pdf_folder):
    if filename.lower().endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder, filename)
        text_path = os.path.join(text_folder, filename.replace(".pdf", ".txt"))

        with fitz.open(pdf_path) as doc:
            text = ""
            for page in doc:
                text += page.get_text()

        with open(text_path, "w", encoding="utf-8") as f:
            f.write(text)

print(" Conversion completed. Text files are saved in 'texts' folder.")
```

```python
import os
import requests
import chromadb
from pathlib import Path
from tqdm import tqdm  # progress bar

# ------------------------
# 1. Settings
# ------------------------
MODEL_NAME = "mistral:latest"              # Model for answering
EMBED_MODEL = "nomic-embed-text"           # Model for embeddings
OLLAMA_API = "http://localhost:11434"
DATA_DIR = Path(r"C:\Users\simon\Desktop\pdf2\research_papers")  # Folder with .txt papers

# Directory where ChromaDB will store the database
CHROMA_DB_DIR = Path(r"C:\Users\simon\Desktop\pdf2\chroma_db")

# ------------------------
# 2. ChromaDB setup (persist to disk)
# ------------------------
chroma_client = chromadb.PersistentClient(path=str(CHROMA_DB_DIR))
collection = chroma_client.get_or_create_collection(name="research_papers")

# ------------------------
# 3. Function to get embeddings from Ollama
# ------------------------
def get_embedding(text: str):
    r = requests.post(f"{OLLAMA_API}/api/embeddings", json={
        "model": EMBED_MODEL,
        "prompt": text
    })
    r.raise_for_status()
    return r.json()["embedding"]

# ------------------------
# 4. Load documents into memory for processing
# ------------------------
all_chunks = []
for file in DATA_DIR.glob("*.txt"):
    text = file.read_text(encoding="utf-8").strip()
    chunks = [text[i:i + 500] for i in range(0, len(text), 500)]
    all_chunks.extend(chunks)

print(f"📄 Found {len(all_chunks)} chunks to embed.")

# ------------------------
# 5. Embed and store with progress bar
# ------------------------
existing_ids = set(collection.get()["ids"])  # Avoid re-embedding duplicates

for idx, chunk in enumerate(tqdm(all_chunks, desc="Embedding chunks", unit="chunk")):
    chunk_id = str(idx)
    if chunk_id in existing_ids:
        continue  # Skip if already embedded
    emb = get_embedding(chunk)
    collection.add(
        ids=[chunk_id],
        documents=[chunk],
        embeddings=[emb]
    )

print("✅ All chunks embedded and stored in ChromaDB.")

# ------------------------
# 6. Query function
# ------------------------
def ask_ollama_with_context(question, top_k=3):
    q_emb = get_embedding(question)
    results = collection.query(query_embeddings=[q_emb], n_results=top_k)
    context = "\n\n".join(results["documents"][0])

    prompt = f"Use the following research context to answer the question:\n\n{context}\n\nQuestion: {question}"

    response = requests.post(f"{OLLAMA_API}/api/generate", json={
        "model": MODEL_NAME,
        "prompt": prompt,
        "stream": False
    })
    return response.json().get("response", "")

# ------------------------
# 7. Example query
# ------------------------
user_question = "Explain the latest advancements in AI from the research papers."
print("\n🤖 Answer:\n", ask_ollama_with_context(user_question))
```

```python
# gradio_rag.py
import os
import json
import requests
import chromadb
import gradio as gr
from pathlib import Path
from tqdm import tqdm

# ------------------------
# 1. Settings 
# ------------------------
MODEL_NAME = "mistral:latest"      
EMBED_MODEL = "nomic-embed-text"           # embedding model name (if Ollama supports it)
OLLAMA_API = "http://localhost:11434"
DATA_DIR = Path(r"C:\Users\simon\Desktop\pdf2\new\texts")  
CHROMA_DB_DIR = Path(r"C:\Users\simon\Desktop\pdf2\new\chroma_db")   # persistent storage

CHUNK_SIZE = 5000

# ------------------------
# 2. ChromaDB setup (persistent)
# ------------------------
# Create persistent client (Chromadb >= certain versions)
chroma_client = chromadb.PersistentClient(path=str(CHROMA_DB_DIR))
collection = chroma_client.get_or_create_collection(name="research_papers")

# ------------------------
# 3. Embedding function (uses Ollama embeddings endpoint)
# ------------------------
def get_embedding(text: str):
    r = requests.post(f"{OLLAMA_API}/api/embeddings", json={
        "model": EMBED_MODEL,
        "prompt": text
    })
    r.raise_for_status()
    return r.json()["embedding"]

# ------------------------
# 4. Helper: create chunks from .txt files
# ------------------------
def load_all_chunks(data_dir: Path, chunk_size: int = CHUNK_SIZE):
    chunks = []
    files = sorted(list(data_dir.glob("*.txt")))
    for file in files:
        text = file.read_text(encoding="utf-8").strip()
        if not text:
            continue
        file_chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
        chunks.extend(file_chunks)
    return chunks

# ------------------------
# 5. Build / update index (called by Gradio button)
#    Uses gr.Progress to show visual progress in the UI.
# ------------------------
def build_index():
    all_chunks = load_all_chunks(DATA_DIR, CHUNK_SIZE)
    total = len(all_chunks)
    if total == 0:
        return "No .txt files found in the research_papers folder."

    # determine existing ids to avoid duplicates (simple scheme: id == index)
    existing_ids = set()
    if collection.count() > 0:
        result = collection.get(limit=collection.count())
        existing_ids = set(result.get("ids", []))

    with gr.Progress(total=total) as progress:
        for idx, chunk in enumerate(all_chunks):
            chunk_id = str(idx)
            if chunk_id in existing_ids:
                progress((idx + 1), desc=f"Skipping {idx+1}/{total} (exists)")
                continue
            try:
                emb = get_embedding(chunk)
            except Exception as e:
                return f"Error getting embedding from Ollama: {e}"
            collection.add(ids=[chunk_id], documents=[chunk], embeddings=[emb])
            progress((idx + 1), desc=f"Embedded {idx+1}/{total}")

    return f"Index built/updated. {collection.count()} total vectors stored."

# ------------------------
# 6. Streaming query function (returns a generator for Gradio streaming)
# ------------------------
def ask_with_context_stream(question: str, top_k: int = 3):
    if not question or question.strip() == "":
        yield "Please enter a question."
        return

    # Get question embedding
    try:
        q_emb = get_embedding(question)
    except Exception as e:
        yield f"Error getting question embedding: {e}"
        return

    # Retrieve top_k docs
    try:
        results = collection.query(query_embeddings=[q_emb], n_results=top_k)
    except Exception as e:
        yield f"Error querying ChromaDB: {e}"
        return

    docs = results.get("documents", [[]])[0]
    if not docs:
        context = ""
    else:
        context = "\n\n".join(docs)

    prompt = (
        "Use the following research context to answer the question concisely and accurately:\n\n"
        f"{context}\n\nQuestion: {question}\nAnswer:"
    )

    # Stream generation from Ollama
    stream_url = f"{OLLAMA_API}/api/generate"
    payload = {"model": MODEL_NAME, "prompt": prompt, "stream": True}

    try:
        with requests.post(stream_url, json=payload, stream=True, timeout=120) as r:
            r.raise_for_status()
            accumulated = ""
            # Ollama sends server-sent-events like lines; parse JSON parts
            for line in r.iter_lines():
                if line:
                    try:
                        text_line = line.decode("utf-8").strip()
                        # many Ollama versions send `data: {...}` SSE lines
                        if text_line.startswith("data:"):
                            json_part = text_line[len("data:"):].strip()
                        else:
                            json_part = text_line
                        # sometimes it sends "[DONE]" or blank
                        if json_part in ("[DONE]", ""):
                            continue
                        # parse as JSON
                        payload_chunk = json.loads(json_part)
                        # common key is 'response' or nested 'choices'
                        chunk_text = ""
                        if isinstance(payload_chunk, dict):
                            # Most simple case
                            chunk_text = payload_chunk.get("response", "")
                            if not chunk_text and "choices" in payload_chunk:
                                # try to parse choices -> delta/content
                                choices = payload_chunk["choices"]
                                if choices and isinstance(choices, list):
                                    # try common fields
                                    for ch in choices:
                                        if isinstance(ch, dict):
                                            chunk_text = chunk_text + ch.get("delta", {}).get("content", "") or ch.get("text", "") or ""
                        if chunk_text:
                            accumulated += chunk_text
                            yield accumulated  # yield accumulated text so UI updates incrementally
                    except json.JSONDecodeError:
                        # ignore unparsable chunks
                        continue
    except Exception as e:
        yield f"Error calling Ollama generate API: {e}"
        return

    yield accumulated  # final output

# ------------------------
# 7. Build Gradio interface
# ------------------------
with gr.Blocks(title="Ollama RAG Demo") as demo:
    gr.Markdown("Research papers Analysis")
    with gr.Row():
        with gr.Column(scale=2):
            q_input = gr.Textbox(label="Question", placeholder="e.g. What is Machine learning?", lines=2)
            top_k = gr.Slider(minimum=1, maximum=8, value=3, step=1, label="Top K retrieved chunks")
            ask_btn = gr.Button("Ask (stream)")
            output = gr.Textbox(label="Model answer (streamed)", lines=10)
        # with gr.Column(scale=1):
        #     build_btn = gr.Button("Build / Update Index")
        #     status = gr.Textbox(label="Status", lines=6)

    # Wire the buttons
    # build_btn.click(fn=build_index, inputs=None, outputs=[status])
    # For streaming, set outputs to the output textbox; Gradio will stream generator yields here
    ask_btn.click(fn=ask_with_context_stream, inputs=[q_input, top_k], outputs=[output])

if __name__ == "__main__":
    demo.launch(server_name="0.0.0.0", server_port=7870, share=False)

```
